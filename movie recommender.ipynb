```python
import nltk 
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk import word_tokenize
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer

def trans(i):
    if i.startswith('J'):
        return wordnet.ADJ
    if i.startswith('R'):
        return wordnet.ADV
    if i.startswith('N'):
        return wordnet.NOUN
    if i.startswith('V'):
        return wordnet.VERB
    else:
        return wordnet.NOUN
    
class Lemmatokenizer:
    def __init__(self):
        self.lemmitizer=WordNetLemmatizer()

    def __call__(self,doc):
        new=word_tokenize(doc)
        pop=nltk.pos_tag(new)
        return [self.lemmitizer.lemmatize(word,trans(tag)) for word,tag in pop]b
        
        
    
```

```python
gen['AllCombiend']=gen['overview']+'  '+gen['stringKeywords']+' ' + gen['StringGenres']
gen
```

```python
vectorizer=TfidfVectorizer(tokenizer=Lemmatokenizer())

output=vectorizer.fit_transform(gen['AllCombiend'])
print(output.shape)
```

```python
vectorofindexTest=10
test1=output[vectorofindexTest]
```

```python
from sklearn.metrics.pairwise import cosine_similarity

sim=cosine_similarity(output,test1)
sim[vectorofindexTest]=0
```

```python

def recommend(name):
    try:
        index = gen.index[gen['original_title'] == name].tolist()[0]
    except IndexError:
        print('the movie is not found')
        return
        
    vectorofindexTest=index
    sim=cosine_similarity(output,output[vectorofindexTest])
    
    data=pd.DataFrame(sim)
    data['names']=gen['original_title']
    data=data.sort_values(by=0,ascending=False)
    first_five_names = data['names'][1:6].tolist()
    
    
    print(first_five_names)
    op=''
    for one in first_five_names:
        op=op+',' +' ' +one
    
    
    return op
```

```python
import tkinter as tk
from tkinter import messagebox

# Function to be executed when button is clicked
def rec(name):
    # Replace this with your actual recommendation logic based on 'name'
    recommendation = recommend(name)
    # Show recommendation in a message box
    messagebox.showinfo("Recommendation", recommendation)

# Function to handle button click and call recommend()
def run_recommendation():
    # Get text from the entry widget (assuming it's for 'name')
    name = entry.get()
    
    # Call recommend() with 'name' as argument
    rec(name)

# Create the main application window
root = tk.Tk()
root.title("Recommendation System")

# Create a label widget for instructions
label = tk.Label(root, text="Enter a name:")
label.pack(pady=10)

# Create an entry widget (input field)
entry = tk.Entry(root, width=30)
entry.pack(pady=10)

# Create a button widget
button = tk.Button(root, text="Get Recommendation", command=run_recommendation)
button.pack(pady=10)

# Start the tkinter event loop
root.mainloop()
```
