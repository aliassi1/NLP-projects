```python
import pandas as pd 
df=pd.read_csv(r"C:\Users\AliOs\Downloads\AirlineTweets.csv")
```

```python
df.head(5)
```

```python
from nltk import word_tokenize
from nltk.stem import WordNetLemmatizer
lemmitizer=WordNetLemmatizer()
pop=[]
for idx,one in enumerate(df['text']):
    sentence=''
    one=word_tokenize(one)
    for word in one:
       
        sentence+=lemmitizer.lemmatize(word)+' '
    pop.append(sentence)
    
        
    
```

```python
outp=[]
for idx,one in enumerate(df['airline_sentiment']):
    if one=='positive':
        x=0
    elif one=='negative':
        x=1
    else:
        x=2
                
    outp.append(x)
    
```

```python
import matplotlib.pyplot as plt
plt.hist(outp)
```

```python
outp=outp[:8000]
```

```python
pop[:8000]
```

```python
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer


vectorizer10=TfidfVectorizer(stop_words=stopwords.words('english'))
```

```python
pop=vectorizer10.fit_transform(pop[:8000])
```

```python
pop
```

```python
from sklearn.decomposition import TruncatedSVD
svd=TruncatedSVD(n_components=8000)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(pop,outp,random_state=42,test_size=0.3,shuffle=True)
```

```python
Z_train=svd.fit_transform(X_train)
Z_Test=svd.transform(X_test)
```

```python


	

def plot_top_words(model, feature_names, n_top_words=5):
  fig, axes = plt.subplots(2, 5, figsize=(30, 15), sharex=True)
  axes = axes.flatten()
  for topic_idx, topic in enumerate(model.components_):
    top_features_ind = topic.argsort()[: -n_top_words - 1 : -1]
    top_features = [feature_names[i] for i in top_features_ind]
    weights = topic[top_features_ind]

    ax = axes[topic_idx]
    ax.barh(top_features, weights, height=0.7)
    ax.set_title(f"Topic {topic_idx +1}", fontdict={"fontsize": 30})
    ax.invert_yaxis()
    ax.tick_params(axis="both", which="major", labelsize=20)
    for i in "top right left".split():
        ax.spines[i].set_visible(False)
    fig.suptitle('LDA', fontsize=40)

  plt.subplots_adjust(top=0.90, bottom=0.05, wspace=0.90, hspace=0.3)
  plt.show()

```

```python
svd.components_
```

```python
feature_names=vectorizer100.get_feature_names_out()
plot_top_words(svd,feature_names)
```
