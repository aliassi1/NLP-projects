```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(df['Poems'],df['Author'],random_state=42,test_size=0.3)
```

```python
import numpy as np 
L=len(wordToIndex)
A0=np.ones((L,L))
pie0=np.ones(L)
A1=np.ones((L,L))
pie1=np.ones(L)
```

```python
X_train
```

```python
def compute_counts(list,A,pi):
    for element in list:
        last_elem = None
        for one in element:
            if last_elem is None:
                pi[one]+=1

            else :
                A[last_elem,one]+=1
                
            last_elem=one
        
        
```

```python
compute_counts([t for t, y in zip(X_train, y_train) if y == 0], A0, pie0)
compute_counts([t for t, y in zip(X_test, y_test) if y == 1], A1, pie1)
```

```python
print(A0)
print('\n')
print(pie0)
```

```python
A0 /= A0.sum(axis=1,keepdims=True)
pie0 /= pie0.sum()

A1 /=A1.sum(axis=1,keepdims=True)
pie1 /= pie1.sum()
```

```python
logA0=np.log(A0)
logpie0=np.log(pie0)

logA1=np.log(A1)
logpie1=np.log(pie1)
```

```python
count0=sum(y==0 for y in y_train)
count1=sum(y==1 for y in y_train)

p0=count0/(count0+count1)

p1= count1/(count0+count1)

print(p0,p1)
logp0=np.log(p0)
logp1=np.log(p1)
```

```python
class Classifier:
    def __init__(self,logpies,logAs,logPrioer):
        self.logAs=logAs
        self.logpies=logpies
        self.logPrioer=logPrioer
        


    def computeProb(self,line,class_):
        logpie=self.logpies[class_]
        logA=self.logAs[class_]
        
        LastIndex=None
        logTotal=0
        
        
        for idx in line:
            if LastIndex is None:
                logTotal+=logpie[idx]

            else :
                logTotal+=logA[LastIndex,idx]

            LastIndex=idx

        return logTotal

    def predict(self,inputs_):
        predict=np.zeros(len(inputs_))
        for idnex,line in enumerate(inputs_):
            out1=[self.computeProb(line,c)+self.logPrioer[c] for c in range(2)]
            out2=np.argmax(out1)
            predict[idnex]=out2

        return predict
        
            
            

        
                
            
```

```python
clf=Classifier([logpie0,logpie1],[logA0,logA1],[logp0,logp1])
```

```python
Ptrain = clf.predict(X_test)
print(f"Train acc: {np.mean(Ptrain == y_test)}")
```

```python
from sklearn.metrics import confusion_matrix,f1_score

y_test_np = y_test.to_numpy()
Ptrain_np = np.array(Ptrain)

# Convert both arrays to integers
y_test_int = y_test_np.astype(int)
Ptrain_int = Ptrain_np.astype(int)

# Compute confusion matrix
cm = confusion_matrix(y_test_int, Ptrain_int)
cm
```
