```python
import numpy as np
N = 100
X = np.random.random(N) * 6 - 3
y = 0.5 * X - 1 + np.random.randn(N) * 0.5
```

```python
import numpy as np
print(X[71])
```

```python
import matplotlib.pyplot as plt 
plt.scatter(X,y)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam




i = Input(shape=(1,))
# this is the input shape

dense=Dense(1)(i)
#the first par is the nb of neurons
#the second one is the input of the layer

model = Model(i, dense)
#the first one is the input of the model
# the second one is the last layer

```

```python
# Display the model summary
model.summary()
#the data input is None in both cases since we might predict 10000 samples at once None acts as a wild card that can take any dim as input 
```

```python
model.compile(
  loss='mse',
  # optimizer='adam',
  optimizer=Adam(learning_rate=0.1),
  metrics=['mae']
)
```

```python
print(X.reshape(-1,1))
```

```python
print(y)
```

```python
r = model.fit(
  X.reshape(-1, 1), y,
  epochs=200,
  batch_size=32,
)
```

```python
r.history['loss']
```

```python
import matplotlib.pyplot as plt 
plt.plot(r.history['loss'], label='loss')
plt.legend();
```

```python
plt.plot(r.history['mae'], label='mae')
plt.legend();
```

```python
#sentiment analysis
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\AliOs\Downloads\AirlineTweets.csv")
```

```python
df
```

```python
df=df[['airline_sentiment','text']]
```

```python
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
```

```python
pop=vectorizer.fit_transform(df['text'])
```

```python
pop.shape[0]/100
```

```python
enco=[]
for one in df['airline_sentiment']:
    if one=='neutral':
        x=1
    elif one=='negative':
        x=2
    else:
        x=0
    enco.append(x)
    
        
```

```python
enco
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(pop,enco,test_size=0.3,random_state=42)
```

```python
from tensorflow.keras.utils import to_categorical
y_train_Hot=to_categorical(y_train)
y_test_Hot=to_categorical(y_test)
```

```python
print(y_test_Hot)
```

```python
print(y_test_Hot[3])
```

```python
X_train
```

```python
inp=Input(shape=(15051,))
```

```python
layer11=Dense(10,activation='relu')(inp)
layer11=Dense(10,activation='relu')(layer11)
layer11=Dense(10,activation='relu')(layer11)

layer11=Dense(3,activation='softmax')(layer11)





```

```python
model11=Model(inp,layer11)
```

```python
model11.summary()
```

```python
from tensorflow.keras.losses import categorical_crossentropy

model11.compile(
    loss='categorical_crossentropy',
    optimizer=Adam(learning_rate=0.001),
    metrics=['accuracy','f1_score','AUC']
    
)
```

```python
import numpy as np
r=model11.fit(
    X_train.toarray(),np.array(y_train_Hot),
    validation_data=(X_test.toarray(), np.array(y_test_Hot)),
    batch_size=130,
    epochs=100
    
)
```

```python
import matplotlib.pyplot as plt
plt.plot(r.history['AUC'])
plt.plot(r.history['val_AUC'])
```

```python
plt.plot(r.history['loss'])
plt.plot(r.history['val_loss'])
```
