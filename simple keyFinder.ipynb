```python
def encode(word):
    word=word.upper()
    str=''
    for char in word:
        str+=enco[char]
    return str
```

```python
def decode(word,dict):
    str=''
    decode_dict={value:key for key,value in dict.items()}
    for char in word:
        str+=decode_dict[char]
    return str
    
```
