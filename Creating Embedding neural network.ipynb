```python
from tensorflow.keras import Input,Model
from tensorflow.keras.layers import Dense,Embedding,Lambda
import tensorflow as tf

context_size=10
input=Input(shape=(10,))
    layers=Embedding(20000,50)(input)
layers=Lambda(lambda t: tf.reduce_mean(t, axis=1))(layers)
layers=Dense(20000,use_bias=False)(layers)

```

```python
model11=Model(input,layers)
```

```python
model11.summary()
```

```python
from tensorflow.keras.losses import CategoricalCrossentropy 


model11.compile(
  loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
  optimizer='adam',
  metrics=['accuracy']
)
```

```python
import numpy as np
import random

half_context_size=context_size//2

def generate_data(output1,batch_size=128):
    X_batch=np.zeros((batch_size,context_size))
    y_batch=np.zeros(batch_size)
    no_of_batches=int(np.ceil(len(output1)/batch_size))
    while True:
        random.shuffle(output1)

        for i in range(no_of_batches):
            
            new_batch=output1[i*batch_size:batch_size*(i+1)]
            len_of_oneBatch=len(new_batch)
    
            for s in range(len_of_oneBatch):
                seq=new_batch[s]
                j = np.random.randint(0, len(seq) - context_size - 1)
    
                x1=seq[j:j+half_context_size]
                x2=seq[j+half_context_size+1:j+context_size+1]
    
                X_batch[s,:half_context_size]=x1
                X_batch[s,half_context_size:]=x2
                y_batch[s]=seq[j+half_context_size]
                
            yield X_batch[:len_of_oneBatch], y_batch[:len_of_oneBatch]
            
        
        
        
        
        
    
    
```

```python
model11.fit(
    generate_data(output1),
    epochs=10000,
    steps_per_epoch=int(np.ceil(len(output1) / 128))
)
```
